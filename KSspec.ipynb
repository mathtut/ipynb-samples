# The Kuramoto-Sivashinsky Equation

The Kuramoto-Sivashinsky equation is computed in this notebook.

It is described as follows:
$$
\begin{align*}
\dfrac{\partial u}{\partial t} + \nu \dfrac{\partial^4 u}{\partial x^4} 
 + \dfrac{\partial^2 u}{\partial x^2} + v \dfrac{\partial u}{\partial x} &= 0 \qquad x \in [0,L]\\\\
\int_0^L u(x) dx &= 0
\end{align*}
$$


In order to use the Fourier spectral method, spatial region is converted to $[0,1]$.
As the result we get

$$
\dfrac{\partial u}{\partial t} + \nu \lambda^4 \dfrac{\partial^4 u}{\partial x^4} 
  + \lambda^2 \dfrac{\partial^2 u}{\partial x^2} + \lambda u \dfrac{\partial u}{\partial x} = 0 
  \qquad x \in [0,1], 
$$

where $\lambda = 1/L$.

Note that you need to "inflate" the number of Fourier modes when calculating the nonlinear term to avoid aliasing errors.

```python
import numpy as np
from scipy.fftpack import rfft, irfft
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def ks(w, t, nu, L):
    lmd = (1.0/L)
    M = len(w)
    J = 4*M
    u = irfft(w, J)
    u *= u
    u *= 0.5  # Now (1/2)*u**2 has been calculated.
    G = rfft(u)
    
    k2pi = 2.0*np.pi*np.arange(1,M/2+1)    
    dG = np.zeros(M)
    
    Anlt = G[1:M:2]
    Bnlt = G[2:M+1:2]
    Anlt *= lmd*k2pi
    Bnlt *= -lmd*k2pi
    
    lnrCoeff = nu*lmd**4*k2pi**4 - lmd**2*k2pi**2
    dG[1::2] = lnrCoeff*w[1::2] + Bnlt  # A'(1), A'(2), ..., A'(M/2)
    dG[2::2] = lnrCoeff[:-1]*w[2::2] + Anlt[:-1] # B'(1), B'(2), ..., B'(M/2 - 1)
    return -dG 
```

```python
nu = 1.0; L = 22.0; T = 200
tarray = np.linspace(0, T, 101)
w = (np.random.rand(64) - 0.5)/10.0
```

```python
wsol = odeint(ks, w, tarray, args=(nu, L), mxstep = 5000)
```

```python
usol = []
for w in wsol:
    u = irfft(w)
    usol.append(u)
usol = np.array(usol)
```

```python
plt.figure(figsize=(8,6))
plt.imshow(usol.T, extent=[0,T,0,L], aspect=4.0)
plt.colorbar(aspect=10, shrink=0.5)
plt.xlabel('t'); plt.ylabel('x')
plt.title('The Kuramoto-Sivashinsky Equation')
```

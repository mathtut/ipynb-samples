# Setting Fonts

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
mpl.rc('font', family='serif', size='18')
```

```python
x = np.linspace(0, 4, 201)
z = 0.2 + 0.5*x
y = np.sin(8.0*np.pi*x)*z
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.plot(x,y, 'b', label='y')
ax.plot(x,z, 'r--', x, -z,'r--')
ax.set_title('$(0.2 + 0.5x)\sin (8\pi x)$')
plt.legend()
```
